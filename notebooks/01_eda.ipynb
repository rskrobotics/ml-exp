# Mining Process EDA

**Goal**: Build a digital twin of a flotation plant to predict concentrate quality (% Silica, % Iron) from process parameters.

## Step 1: Load and Inspect the Data

First thing in any ML project: understand what you're working with.
- How big is the data?
- What are the column types?
- Any missing values?

```python
import pandas as pd
import numpy as np

# Load data - note the European decimal format (comma instead of dot)
df = pd.read_csv(
    '../data/raw/MiningProcess_Flotation_Plant_Database.csv',
    decimal=',',  # Handle European number format
    parse_dates=['date']
)

print(f"Shape: {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1e6:.1f} MB")
print(f"\nTime range: {df['date'].min()} to {df['date'].max()}")
```
